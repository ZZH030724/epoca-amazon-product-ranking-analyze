```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# 1) ж•°жҚ®и·Ҝеҫ„пјҢиҜ·жҢүе®һйҷ…зҺҜеўғи°ғж•ҙ
base_path = "/blue/qmb6938/_Epoca/Amazon_Rankings/cleaned_auto"  # дҝ®ж”№дёәдҪ зҡ„зӣ®еҪ•
file_name = "enhanced_amazon_rankings.csv"                   # ж–Үд»¶еҗҚеҰӮдҪ жүҖиҝ°
path = os.path.join(base_path, file_name)

# 2) иҜ»еҸ–ж•°жҚ®
df = pd.read_csv(path)

# 3) иҮӘеҠЁйҖүеҸ–з”ЁдәҺж ҮжіЁзҡ„еҲ—пјҲtitle/itle жҲ– brand_nameпјү
label_cols = ["title", "itle", "brand_name"]
label_col = next((c for c in label_cols if c in df.columns), None)
if label_col is None:
    # еҰӮжһңйғҪдёҚеӯҳеңЁпјҢдҪҝз”Ёз¬¬дёҖеҲ—дҪңдёәж Үзӯҫ
    label_col = df.columns[0]

# 4) зЎ®дҝқ rank дёәж•°еҖјеһӢпјҢеҺ»йҷӨзјәеӨұзҡ„еҝ…иҰҒдҝЎжҒҜ
df['rank'] = pd.to_numeric(df['rank'], errors='coerce')
df = df.dropna(subset=['category', 'rank', label_col])

# 5) жҢү category еҲҶз»„пјҢйҖүеҸ–жҺ’еәҸеҗҺзҡ„еүҚ top_n жқЎпјҲиҝҷйҮҢи®ҫдёә 3пјҢеҸҜд»Ҙж”№жҲҗд»»ж„Ҹж•ҙж•°пјү
top_n = 3
df_top = df.sort_values(['category', 'rank']).groupby('category').head(top_n)

# 6) еҸҜи§ҶеҢ–пјҡдёәжҜҸдёӘ category иҫ“еҮәдёҖдёӘеӯҗеӣҫпјҢжЁӘеҗ‘жқЎеҪўиЎЁзӨә rankпјҢж Үзӯҫдёәдә§е“Ғ/е“ҒзүҢеҗҚ
categories = df_top['category'].unique()
ncols = min(4, len(categories))  # жҜҸиЎҢжңҖеӨҡ 4 еҲ—еӯҗеӣҫпјҢзұ»еҲ«еӨӘеӨҡж—¶еҸҜжү©еұ•
fig, axes = plt.subplots(1, ncols, figsize=(5 * ncols, 6), squeeze=False)

for ax, cat in zip(axes.flat, categories[:ncols]):
    sub = df_top[df_top['category'] == cat].sort_values('rank')
    labels = sub[label_col].astype(str).values
    ranks = sub['rank'].values
    ax.barh(labels, ranks, color='steelblue')
    ax.set_title(cat)
    ax.set_xlabel('Rank (lower is better)')
    ax.invert_yaxis()

plt.tight_layout()
out_dir = "./visualizations_enhanced_by_category"
os.makedirs(out_dir, exist_ok=True)
plt.savefig(os.path.join(out_dir, "top3_by_category_barplots.png"))
plt.close()

print("еҸҜи§ҶеҢ–е·Іе®ҢжҲҗпјҢиҫ“еҮәзӣ®еҪ•:", out_dir)
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def main():
    # ж•°жҚ®и·Ҝеҫ„пјҢиҜ·жҢүе®һйҷ…зҺҜеўғи°ғж•ҙ
    base_path = "/blue/qmb6938/_Epoca/Amazon_Rankings/cleaned_auto"  # дҝ®ж”№дёәдҪ зҡ„зӣ®еҪ•
    file_name = "enhanced_amazon_rankings.csv"                   # ж–Үд»¶еҗҚ
    path = os.path.join(base_path, file_name)

    # иҜ»еҸ–ж•°жҚ®
    df = pd.read_csv(path)

    # йҖүжӢ©з”ЁдәҺж ҮжіЁзҡ„еҲ—пјҲtitle/itle жҲ– brand_nameпјү
    label_cols = ["title", "itle", "brand_name"]
    label_col = next((c for c in label_cols if c in df.columns), None)
    if label_col is None:
        label_col = df.columns[0]

    # еӨ„зҗҶж—ҘжңҹеҲ—
    if "date" in df.columns:
        df["date"] = pd.to_datetime(df["date"], errors='coerce')
        df = df.dropna(subset=["date", "trend_score", "volatility"])
        df["month"] = df["date"].dt.to_period("M").dt.to_timestamp()
    else:
        print("иӯҰе‘Ҡпјҡж•°жҚ®дёӯжІЎжңү date еҲ—пјҢж—¶й—ҙеәҸеҲ—еҸҜи§ҶеҢ–е°Ҷж— жі•жҢүж—¶й—ҙеҲҶж®өгҖӮ")
        # еҰӮжһңжІЎжңүж—ҘжңҹеҲ—пјҢеҗҺз»ӯж—¶й—ҙеәҸеҲ—е°Ҷж— жі•жӯЈзЎ®иҒҡеҗҲпјҢжӯӨеӨ„зӣҙжҺҘйҖҖеҮәжҲ–ж”№жҲҗе…ЁеұҖиҒҡеҗҲ
        return

    # ж•°еҖјеҲ—иҪ¬жҚўпјҢзЎ®дҝқ trend_scoreгҖҒvolatility дёәж•°еҖјеһӢ
    df["trend_score"] = pd.to_numeric(df["trend_score"], errors="coerce")
    df["volatility"] = pd.to_numeric(df["volatility"], errors="coerce")
    df = df.dropna(subset=["category", "trend_score", "volatility", "month", label_col])

    # йҖүеҸ–еүҚиӢҘе№Ізұ»еҲ«иҝӣиЎҢеҸҜи§ҶеҢ–пјҢйҒҝе…ҚзәҝеӨӘеӨҡзңӢдёҚжё…
    top_n = 6
    top_categories = df["category"].value_counts().nlargest(top_n).index.tolist()
    df_top = df[df["category"].isin(top_categories)].copy()

    out_dir = "./visualizations_enhanced_time_series"
    os.makedirs(out_dir, exist_ok=True)

    # еҸҜи§ҶеҢ– 1пјҡtrend_score зҡ„ж—¶й—ҙеәҸеҲ—пјҲжҢүжңҲгҖҒжҢүзұ»еҲ«пјү
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=df_top, x="month", y="trend_score", hue="category", estimator="mean", ci=None)
    plt.title("Monthly Mean Trend Score by Category (top {} categories)".format(top_n))
    plt.xlabel("Month")
    plt.ylabel("Trend Score")
    plt.legend(title="Category", bbox_to_anchor=(1.05, 1), loc="upper left")
    plt.tight_layout()
    plt.savefig(os.path.join(out_dir, "trend_score_by_category_time_series.png"))
    plt.close()

    # еҸҜйҖүеҸҜи§ҶеҢ– 2пјҡvolatility зҡ„ж—¶й—ҙеәҸеҲ—пјҲжҢүжңҲгҖҒжҢүзұ»еҲ«пјү
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=df_top, x="month", y="volatility", hue="category", estimator="mean", ci=None, palette="tab10")
    plt.title("Monthly Mean Volatility by Category (top {} categories)".format(top_n))
    plt.xlabel("Month")
    plt.ylabel("Volatility")
    plt.legend(title="Category", bbox_to_anchor=(1.05, 1), loc="upper left")
    plt.tight_layout()
    plt.savefig(os.path.join(out_dir, "volatility_by_category_time_series.png"))
    plt.close()

    print("ж—¶й—ҙеәҸеҲ—еҸҜи§ҶеҢ–е·Іе®ҢжҲҗпјҢиҫ“еҮәзӣ®еҪ•:", out_dir)

if __name__ == "__main__":
    main()
```

```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def visualize_kitchen_only(base_path, file_name="enhanced_amazon_rankings.csv", top_n=10, out_dir="visualizations_kitchen"):
    path = os.path.join(base_path, file_name)

    # Load data
    df = pd.read_csv(path)

    # Detect label column for plotting
    label_cols = ["title", "itle", "brand_name"]
    label_col = next((c for c in label_cols if c in df.columns), None)
    if label_col is None:
        label_col = df.columns[0]
    df[label_col] = df[label_col].astype(str)

    # Ensure rank is numeric
    if "rank" in df.columns:
        df["rank"] = pd.to_numeric(df["rank"], errors="coerce")
    else:
        raise SystemExit("Missing 'rank' column")

    # Filter Kitchen category
    df_k = df[
        df["category"].astype(str).str.contains("Kitchen", case=False, na=False) |
        (df["category"].astype(str) == "Kitchen & Dining")
    ]

    if df_k.empty:
        print("No kitchen category data found.")
        return

    os.makedirs(out_dir, exist_ok=True)

    # 1) Top items by average rank within Kitchen
    grp = df_k.groupby(label_col)["rank"].mean().sort_values()
    top_label = grp.head(top_n).index.tolist()

    plt.figure(figsize=(8, max(4, top_n*0.5)))
    top_df = df_k[df_k[label_col].isin(top_label)].copy()
    top_df = top_df.sort_values("rank")
    sns.barplot(x="rank", y=label_col, data=top_df, orient="h", palette="Blues_d")
    plt.title(f"Top {top_n} Kitchen Products by Rank (lower is better)")
    plt.xlabel("Rank (lower is better)")
    plt.tight_layout()
    plt.savefig(os.path.join(out_dir, "kitchen_top_by_rank.png"))
    plt.close()

    # 2) Time series of trend_score and volatility for top labels
    if "date" in df_k.columns:
        df_k["date"] = pd.to_datetime(df_k["date"], errors="coerce")
        df_k = df_k.dropna(subset=["date", "trend_score", "volatility"])
        df_k["month"] = df_k["date"].dt.to_period("M").dt.to_timestamp()

        # choose top 5 by total occurrences
        top_labels = df_k[label_col].value_counts().index[:5]
        ts = df_k[df_k[label_col].isin(top_labels)].groupby([df_k["month"], label_col]).agg(
            Trend_Score=("trend_score","mean"),
            Volatility=("volatility","mean")
        ).reset_index()

        plt.figure(figsize=(12,6))
        sns.lineplot(data=ts, x="month", y="Trend_Score", hue=label_col)
        plt.title("Monthly Mean Trend Score for Top Kitchen Items")
        plt.xlabel("Month")
        plt.ylabel("Trend Score")
        plt.tight_layout()
        plt.savefig(os.path.join(out_dir, "kitchen_trend_time_series.png"))
        plt.close()

        plt.figure(figsize=(12,6))
        sns.lineplot(data=ts, x="month", y="Volatility", hue=label_col)
        plt.title("Monthly Mean Volatility for Top Kitchen Items")
        plt.xlabel("Month")
        plt.ylabel("Volatility")
        plt.tight_layout()
        plt.savefig(os.path.join(out_dir, "kitchen_volatility_time_series.png"))
        plt.close()
    else:
        print("No date column found; skipping time-series visualization.")

    # 3) Optional: scatter plot of Trend vs Volatility for kitchen items
    if "trend_score" in df_k.columns and "volatility" in df_k.columns:
        sample = df_k.dropna(subset=["trend_score", "volatility"])
        if len(sample) > 0:
            plt.figure(figsize=(6,4))
            sns.scatterplot(x="volatility", y="trend_score", data=sample, alpha=0.6)
            plt.title("Trend Score vs Volatility (Kitchen Category)")
            plt.xlabel("Volatility")
            plt.ylabel("Trend Score")
            plt.tight_layout()
            plt.savefig(os.path.join(out_dir, "kitchen_trend_vs_volatility.png"))
            plt.close()

    print("Kitchen visualization completed. Output: ", out_dir)

if __name__ == "__main__":
    base_path = "/blue/qmb6938/_Epoca/Amazon_Rankings/cleaned_auto"
    visualize_kitchen_only(base_path, file_name="enhanced_amazon_rankings.csv", top_n=10)
```
